```python
import pandas as pd
df=pd.read_csv("hotel_reviews.csv")
```

```python
df
```

```python
newdf=df.drop(columns=["Area","Review_Date","Rating(Out of 10)"])
newdf
```

```python
newdf["Name"]=newdf["Name"] + ", " + newdf["Rating_attribute"] + ", " + newdf["Review_Text"]
newdf['Name']
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
```

```python
embeddings=HuggingFaceEmbeddings()
```

```python
from langchain.schema import Document
documents = []
for index, row in newdf.iterrows():
    text = str(row['Name']) + " " + str(row['Rating_attribute']) + " " + str(row['Review_Text'])
    doc = Document(page_content=text, metadata={"Name": row['Name'], "Rating_attribute": row['Rating_attribute'], "Review_Text":row["Review_Text"]})
    documents.append(doc)
documents[0]
print(len(documents))
```

```python
documents[0]
```

```python
vector_store=FAISS.from_documents(documents, embeddings)
```

```python
retriever=vector_store.as_retriever()
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.chains import RetrievalQA

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=50
)

gpt2_model = HuggingFacePipeline(pipeline=pipe)

qa_chain = RetrievalQA.from_chain_type(
    llm=gpt2_model,
    retriever=retriever,
    chain_type="stuff",
    return_source_documents=True
)

query = "What's the review about Hotel The Pearl"

response = qa_chain({"query": query})
print(response['result'])
```
